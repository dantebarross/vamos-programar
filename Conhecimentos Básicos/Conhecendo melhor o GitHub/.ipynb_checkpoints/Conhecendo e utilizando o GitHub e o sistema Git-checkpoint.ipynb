## 1. Git e GitHub, são a mesma coisa? <a name="git1"></a>

Se você já se deparou com os temos **Git** e **GitHub**, talvez tenha percebido que eles não são a mesma coisa.

O **Git** é um software utilizado para controlar a versão (_Sistema de Controle de Versão_, ou VCS) de um código ou de projetos inteiros. Pode ser utilizado individualmente ou em equipe, através do terminal ou da interface. Exemplo: você já está na terceira versão (3.0) de um programa, e utilizou o Git em todas as etapas de programação. O código da versão 1.0 precisa ser resgatado? Com o Git você **pode** realizar esse resgate, o que torna o desenvolvimento do projeto muito mais seguro, além de outras vantagens como o trabalho em equipe, todos ao mesmo tempo em um mesmo projeto! É uma questão de organização, saber quem fez o que, o que mudou, nos mínimos detalhes. Se a mudança ocasionou algum bug, o retorno a uma versão anterior pode ser feito. Para que o sistema dê mais certo ainda, as mudanças podem e devem ser comentadas.

Já o **GitHub** é uma plataforma que procura integrar o sistema acima (ser um repositório de projetos no estilo VCS), mas sua intenção vai além disso! Aqui no GitHub, existe a proposta de disponibilizar seu código para o mundo, reunir programadores para que ajudem a melhorar projetos e também como rede social. As formas de utilizar o GitHub são diversas. Aqui, por exemplo, tento passar um pouco do meu conhecimento para quem deseja realizar o básico e um pouco mais, além de compreender cada etapa no processo de desenvolvimento. O GitHub pode inclusive ser visto como uma espécie de **LinkedIn dos programadores**. É isso mesmo! Aqui você exibe seus projetos, fala um pouco mais de si mesmo e pode ser contatado a qualquer instante por recrutadores.

## 2. Mas como funciona o sistema do Git? <a name="git2"></a>

Aqui, nós vamos falar sobre a parte de Versionamento desses sistemas, como funciona e como podem ser utilizados.

Utilizando o sistema do Git, o processo que acontece pode ser simplificado abaixo:
1. Criação de um **repositório**, seja no seu próprio computador ou em Nuvem (no GitHub, por exemplo);
2. Ao desenvolver um código, realizar o **commit** (uma equipe toda pode realizar seus commits nesse repositório, há a possibilidade de adicionar e restringir acessos). Ao realizar um commit, você fará o armazenamento daquela versão do código ao seu repositório local (normalmente em sua máquina). Também serão armazenadas informações de **quem** fez o commit, **o que** adicionou/removeu/substituiu, **quando** fez isso e **quais** os comentários do commit fez, se houver.
3. Realizar o push da versão de seu código e informações para o repositório remoto (só aqui as alterações estarão disponíveis para a equipe), ou seja, para o repositório da equipe (**que pode muito bem estar no GitHub**).

Um bom exemplo para entender a complexidade e vantagem de um sistema de versionamento como o Git:
Se você e outros colegas estão trabalhando no mesmo código, primeiramente devem se comunicar. Depois de realizadas as alterações, ao realizar o commit e o push, o Git fará com que os conflitos sejam os mínimos possíveis. Se você alterou apenas o final do código, ou adicionou comentários, somente essas alterações serão consideradas. Caso seu colega venha a dar commit e push depois disso, o código dele será adicionado apenas onde foi alterado. O resultado final é: as alterações dele e as suas alterações, juntas! Se vocês alteraram a mesma linha de código, as duas alterações serão mostradas, provavelmente como forma de alerta, em uma cor diferente, através de uma notificação, etc.

É claro, devemos evitar ao máximo tais mudanças! Futuramente talvez tenhamos uma plataforma Git com alterações ao vivo, para que duas ou mais pessoas trabalhem na exata mesma versão de código, com commits e comentários em equipe, rodando em uma espécie de repositório compartilhado.


## 3. Tudo bem, e onde entra o GitHub nisso? <a name="git3"></a>
Se você usou o GitHub para armazenar seus programas, percebeu que ele serve como repositório de armazenamento (público e privado), e também que é muito melhor do que o Google Drive ou qualquer outro serviço de nuvem, pois além de armazenar, o GitHub também exibe o código em todas as linguagens de programação do mercado.

Além de servir de repositório de armazenamento e de todas as vantagens citadas no começo desse documento, o GitHub **pode ser utilizado em conjunto com o sistema Git**, onde o **Git** servirá como seu **repositório local**, (sua máquina) e o **GitHub** será o **repositório remoto**, funcionando em nuvem e possibilitando o trabalho em equipe dentro de uma plataforma muito amigável, inclusive possibilitando que pessoas de fora contribuam com sugestões e com alterações no código. Os repositórios públicos do GitHub são **ilimitados**. Os privados também, porém aceitam apenas três ou menos colaboradores.

Se por acaso não curtiu o GitHub por qualquer aspecto, existem outros que funcionam de forma muito parecida (como um repositório remoto para trabalhar em equipe através do Git):
1. GitLab - https://about.gitlab.com/
2. Kallithea - https://kallithea-scm.org/
3. Gogs - https://gogs.io/
4. Phabricator - https://www.phacility.com/phabricator/
5. Bitbucket - https://bitbucket.org/

Minha experiência pessoal foi apenas com o GitLab, e não tenho reclamações. A gente sabe como os culpados por erros geralmente são de quem está atrás das telas, rsrsrs.

## 4. Instalando o Git, o GitHub Desktop e o IDE (Windows) <a name="git4"></a>
Para instalar o sistema Git, basta acessar o site oficial na sessão downloads e escolher seu SO: https://git-scm.com/downloads

Uma das maneiras mais amigáveis de se utilizar o Git é através dos GUI Clients, ou seja, softwares visuais, clicáveis. Sem eles, o Git funciona através da Command Line/Terminal de seu computador.

No site oficial do Git, na sessão GUI Clients, podemos encontrar o GitHub Desktop. Por enquanto só possui suporte para Windows e Mac, mas você também pode procurar outros GUI Clients para seu Linux (**na sessão do Linux, ensinaremos um outro jeito de instalar o GitHub Desktop!**)

Sessão das GUI Clients - https://git-scm.com/downloads/guis
Site oficial do GitHub Desktop - https://desktop.github.com/

**ATENÇÃO!!**: Antes de instalar o GitHub Desktop, se você não tiver uma conta GitHub, cadastre-se na plataforma!

Podemos reparar na interessante definição do GitHub Desktop em seu site oficial:

"Focus on what matters instead of fighting with Git. Whether you're new to Git or a seasoned user, GitHub Desktop simplifies your development workflow"

Você também precisará de um editor de códigos. Pode ser até mesmo o Jupyter Notebook, porém para ter uma melhor integração com o GitHub, utilizaremos uma IDE que já tenha **controle Git incorporado**.

O **PyCharm** é o meu IDE preferido, e sempre recomendo! Olharemos atentamente neste guia para ele e para o **VS Code**:

Baixe a versão Community do PyCharm - https://www.jetbrains.com/pt-br/pycharm/download/

Baixe a versão Stable do VS Code - https://code.visualstudio.com/

## 5. Instalando o Git, o GitHub Desktop e o IDE (Linux) <a name="git5"></a>
O Git foi originalmente feito para Linux. Isso mesmo! Linus Torvalds, **o próprio criador do Linux**, desenvolveu também o **Git** por conta de uma **briga feia**. Pra saber mais, você pode acessar a história no Wikipedia https://pt.wikipedia.org/wiki/Git.


Aberta sua distribuição Linux, você provavelmente já tem acesso ao Terminal e à senha de super usuário (sudo). Abra-o e realize os comandos abaixo:

``
sudo apt update
sudo apt upgrade -y
sudo apt install git -y
``


**DICAS!**:
- O "-y" faz com que não precisemos ficar confirmando com o "y" de _yes_ toda hora.
- Para checar sua versão git, utilize o comando **git --version**.

Se ainda não tem, instale um IDE para editarmos os códigos e acessarmos o projeto. Recomendo os dois abaixo.
1. A versão Community do PyCharm - https://www.jetbrains.com/pt-br/pycharm/download/
2. A versão Stable do VS Code - https://code.visualstudio.com/

No Linux ainda não é possível instalar o software GitHub Desktop diretamente do site oficial (mas veremos outro método **agora**). 
Para encontrar o instalador, visitaremos o repositório **desktop** do usuário shiftkey, criador desse método: https://github.com/shiftkey/desktop

O arquivo de instalação está no link a seguir, página dos _Last releases_: https://github.com/shiftkey/desktop/releases

Clique na última versão (_release_) e escolha o arquivo de instalação de sua distribuição Linux (se você não souber qual é, arrisque o arquivo _.deb_, para as distribuições baseadas no Debian).

A instalação deu certo? Siga o passo a passo na sessão a seguir!

## 6. GitHub Desktop e criando um repositório <a name="git6"></a>
No processo de instalação do GitHub Desktop, faça o _Sign In_ em sua conta do GitHub.

Ao abrir o software, já nos deparamos com algumas opções das mais utilizadas:

<img src="images/options github desktop.png"  alt="GitHub Desktop"/>

As opções são:
1. Criar um repositório tutorial, onde você passará por um rápido guia para aprender as principais funções
2. Clonar um repositório da Internet (pode ser seu ou até mesmo de outra pessoa!)
3. Criar um novo repositório em seu HD, com isso você criará uma pasta no computador que será associada a um repositório no GitHub.
4. O mesmo que antes, porém agora você pode escolher uma pasta, independente de estar vazia ou com projetos.

Vou criar o repositório em que será inserido *esse* guia, pois estava aguardando esse momento! O padrão é criar um repositório privado, que só você poderá acessar (ou quem tiver permissão no futuro).

Ao clicar em  necessário dar um nome a ele, descrição, inserir a pasta do projeto e uma licença. Recomendo escolher a licença MIT.

Acabamos de criar um **repositório local**, ou seja, ele ainda não aparece em seu repositório online no site do GitHub. Para que isso aconteça, precisamos escolher a opção **Publish repositório**.

## 7. Utilizando o GitHub Desktop e os IDEs <a name="git7"></a>
Repositório criado e publicado no GitHub. Ao baixar e instalar o VS Code ou o PyCharm, clique na aba _Files_ (ou _Arquivo_) e vá em _Open Folder_ (ou _Abrir Pasta_). Selecione a pasta que foi criada dentro do GitHub Desktop. O seu **repositório local** será reconhecido e aberto dentro da IDE. Podemos acessar as mudanças e também as diferentes _branchs_ de seu repositório nas setas mostradas abaixo (no VS Code).

<img src="images/ide git.png"  alt="IDE Git" width=200/>


Clicando na opção das três bolinhas, você perceberá que o **VS Code** consegue realizar commits e várias outras funções incluindo o push, direto do IDE! Clique em alguma das opções e uma janela para login no GitHub será aberta. É uma opção viável.

Ao abrir o repositório local no **PyCharm**, ele também detectará o sistema Git. Realize o commit de suas alterações e clique na opção **Push**. Uma janela para logar no GitHub aparecerá. Desse modo, **você não precisa acessar o GitHub Desktop toda vez que quiser realizar um commit, ou push, ou trocar de branch**.

## 8. Fazendo o commit e push de alterações <a name="git8"></a>

Na imagem abaixo, podemos observar que eu realizei a correção de pontuação, salvei o arquivo e, ao clicar em _Fetch origin_ dentro do meu repositório no GitHub Desktop, a pequena alteração foi localizada!

<img src="images/alteracao_fetch.png"  alt="Alteração no Git" width=600/>

Para confirmar as alterações, podemos criar em Commit. Caso queiramos descartar quaisquer alterações, ou ignorar o arquivo para que suas alterações não sejam notificadas nem Commitadas pelo Git, clique com o botão direito no arquivo desejado. As opções estão indicadas na imagem abaixo.

<img src="images/commit_or_not.png"  alt="Commit Git" width=250/>

Para "commitar", escolha um título da alteração e uma descrição do que foi feito. O servidor remoto do GitHub só receberá as alterações após ser feito o **Push**. Para isso, no GitHub Desktop clique no botão **Push origin** ou aperte _Ctrl + P_.

Voilà! Suas alterações agora podem ser vistas no repositório online. Quaisquer alterações que forem feitas em seu projeto no repositório local (na pasta de seu computador) aparecerão como alterações. Você pode inclusive utilizar o Jupyter Notebook, cujo tutorial já temos aqui em meu repositório de guias.

Você fez algum alteração diretamente no repositório online do GitHub ou seu colega também fez alguma alteração? **Elas só serão exibidas quando atualizar o repositório através do botão _Fetch origin_ já visto anteriormente**. Se algo foi alterado e commitado no repositório online e seu repositório local ainda não possuir a alteração, o Git acusará isso e, para realizar a atualização, clique no botão _Pull origin_ ou aperte _Ctrl + Shift + P_. Assim seu repositório local estará exatamente igual o repositório online.

## 9. Clonando um repositório já existente <a name="git9"></a>
Se um repositório está público (ou você tem acesso a ele mesmo privado), existe algumas formas de clona-lo, transferindo para uma pasta de seu computador em formato de repositório local. Ao utilizar sua IDE e o GitHub Desktop, talvez já tenha se deparado com a opção Clonar Repositório, vamos fazer isso através do GitHub Desktop.

Para isso, primeiro você deve abrir o repositório que deseja clonar através do GitHub (em outros sistemas como o GitLab também existe essa possibilidade! Ambos são sistemas Git). 

Clique no botão indicado abaixo. Todas as opções são válidas para fazer o download do repositório, porém para clona-lo como um sistema Git, não basta apenas realizar o download em Zip. Clique em **Open with GitHub Desktop**. Aceite a permissão se for notificado!


<img src="images/clone repository.png"  alt="Clone Repository" width=700/>

Uma janela contendo a URL do repositório será aberta, bem como o path que pode ser alterado.

Outro jeito de fazer isso dentro do GitHub Desktop é o de clicar em _File_ e em _Clone Repository_ ou apertar _Ctrl + Shift + O_. Lá, você seleciona a pasta desejada e cola a URL do repositório GitHub.

Pronto! Você já pode abrir a pasta no seu IDE, rodar os códigos do programa, acessar e modificar arquivos e subir do seu repositório local para seu repositório online do GitHub. **ATENÇÃO: NUNCA FALE QUE É SEU, O QUE NÃO É SEU**.

Clonar o repositório de alguém, modificá-lo e realizar o push para o seu próprio repositório não significa que o programa agora é seu. Se tem objetivo de melhorar o código, cite o criador, a URL do repositório original e informe o que foi alterado ao publicar.

Lembre-se que **um programa pode ser de código aberto, mas a propriedade intelectual de quem o criou existe e o crédito deve ser dado, incluindo todos seus códigos, programas e repositórios**!
